# K-Nearest Neighbors (K-NN)

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values #all rows and columns except last coloumn
y = dataset.iloc[:, -1].values
#dataset
#i.loc means integer location
X
#print(y)
#print(dataset)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1)
```

## Feature Scaling

```python
#from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
#sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Training the K-NN model on the Training set

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 3, metric = 'minkowski', p = 1)
classifier.fit(X_train, y_train)
```

## Predicting a new result

```python
print(classifier.predict(sc.transform([[55,125000]]))) # predicting a result for a customer whose age is 55 and income is 125000
print(classifier.predict(sc.transform ([[26,80000]])))
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
# here we are comparing the predicted label and actual label of the test records one be one
#y_pred is predicated label by the model and y_test is actaul label of the test record prsent in dataset
#y_pred
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score,  precision_score,recall_score, f1_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
precision_score(y_test, y_pred)
#recall_score(y_test, y_pred)
#f1_score(y_test, y_pred)
```
